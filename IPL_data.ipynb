```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the dataset
df_IPL = pd.read_csv('IPL_Dataset.csv')
```

```python
# first 5 rows of the dataframe
df_IPL.head()
```

```python
#Q1. Maximum number of matches played by an individual player in a season:
max_matches = df_IPL['Mat'].max()
print(f'Maximum number of matches played by an individual player in a season: {max_matches}')
```

```python
#Q2. Highest average of a player in a season:
highest_average = df_IPL['Avg'].max()
print(f'Highest average of a player in a season: {highest_average}')
```

```python
#Q3. Average number of matches played by the players in a season:
average_matches = df_IPL['Mat'].mean()
print(f'Average number of matches played by the players in a season: {average_matches}')
```

```python
#Q4. Please check the correlation between all the features:
# Ensure only numeric columns are included for correlation
numeric_data = df_IPL.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()
correlation_matrix
```

```python
#Q5. Check the list of players who have an average greater than 50 and a strike rate above 120.

Players_avg50_sr120 = df_IPL[(df_IPL['Avg'] > 50) & (df_IPL['SR'] > 120)][['Player','Avg','SR']]
Players_avg50_sr120
```

```python
 #Q6. Please check the list of players who have an average greater than 40 and balls faced above 100.

players_avg40_bf100 = df_IPL[(df_IPL['Avg'] > 40) & (df_IPL['BF'] > 100)][['Player','Avg','BF']]
players_avg40_bf100
```

```python
#Q7. Please check the list of players who scored more than 300 runs with an average greater than 45 & strike rate above 130.

df_IPL[(df_IPL['Runs'] > 300) & (df_IPL['Avg'] > 45) & (df_IPL['SR'] > 130)][['Player','Runs','SR','Avg']]
```

```python
#Q8. Players who scored at least one century in this season.

df_IPL[df_IPL['100'] > 0][['Player','100']]
```

```python
#Q9. Players who scored at least 4 half-centuries in this season.

df_IPL[df_IPL['50']>=4][['Player','50']]
```

```python
#Q10. Check the list of players who hit more than 45 boundaries and more than 10 sixes in this season.

df_IPL[(df_IPL['4s']>45) & (df_IPL['6s']>10)][['Player','4s','6s']]
```

```python
#Q11. Plot a histogram of the number of matches played in a season by players.

import matplotlib.pyplot as plt

plt.hist(df_IPL['Mat'], bins=range(df_IPL['Mat'].min(), df_IPL['Mat'].max() + 1), edgecolor='black')
plt.xlabel('Number of Matches')
plt.ylabel('Frequency')
plt.title('Histogram of Number of Matches Played in a Season by Players')
plt.show()
```

```python
#Q12. Plot the histogram of balls faced by players.

plt.hist(df_IPL['BF'], bins=20, edgecolor='black')
plt.xlabel('Balls Faced')
plt.ylabel('Frequency')
plt.title('Histogram of Balls Faced by Players')
plt.show()
```

```python
#Q13. Top 20 players with the most runs in a season.

df_IPL.nlargest(20, 'Runs')[['Player','Runs']]
```

```python
#Q14. Bar chart for top 20 players with the best strike rate in a season.

Top_20_players_sr = df_IPL.nlargest(20, 'SR')
plt.bar(Top_20_players_sr['Player'], Top_20_players_sr['SR'])
plt.xlabel('Players')
plt.ylabel('Strike Rate')
plt.title('Top 20 Players with Best Strike Rate in a Season')
plt.xticks(rotation=90)
plt.show()
```

```python
#Q15. Top 30 players with the highest number of sixes.

df_IPL.nlargest(30, '6s')[['Player','6s']]
```

```python
#Q16. Scatter plot of runs scored by a player vs. balls faced in a season.

plt.scatter(df_IPL['BF'], df_IPL['Runs'])
plt.xlabel('Balls Faced')
plt.ylabel('Runs Scored')
plt.title('Runs Scored vs. Balls Faced in a Season')
plt.show()
```

```python
#Q17. KDE plot for averages in a season.

import seaborn as sns

sns.kdeplot(df_IPL['Avg'])
plt.xlabel('Average')
plt.title('KDE Plot for Averages in a Season')
plt.show()
```
